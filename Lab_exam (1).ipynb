Task 1. Drop words if not alphabats


```python
import re

def remove(news):
    news = re.sub(r'[^a-zA-Z\s]', '', news, re.I|re.A)
    news = re.sub(' +', ' ', news)
    return news


df['cleaned_text'] = df['news'].apply(remove)
df.head()
```

Task 2. Tokenize the sentences

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize
```

```python
def tokenize(clean_text):
    tokens = word_tokenize(clean_text)
    return tokens

df['tokenized_text'] = df['cleaned_text'].apply(tokenize)
df.head()
```

Perform lemitization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()
def lemmitize_words(text):
    return " ".join([lemmitizer.lemmatize(word,pos='v') for word in text.split()])
```

```python
# df['cleaned_text'][4]
```

```python
df['lemat'] = df['cleaned_text'].apply(lemmitize_words)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
X= df['lemat']
Y= df['type']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
bi_gram_vectorizer = CountVectorizer(ngram_range=(2,2))

X_train_n_gram = bi_gram_vectorizer.fit_transform(X_train)
X_test_n_gram = bi_gram_vectorizer.transform(X_test)

# Output the shapes of the resulting Bag of Words matrices
print(f"Shape of X_train_bow: {X_train_n_gram.shape}")
print(f"Shape of X_test_bow: {X_test_n_gram.shape}")

rf = RandomForestClassifier(n_estimators=150)

rf.fit(X_train_n_gram,y_train)
y_pred = rf.predict(X_test_n_gram)

print (accuracy_score(y_test,y_pred))
print (confusion_matrix(y_test,y_pred))
```
